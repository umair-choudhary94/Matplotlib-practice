```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,52,54,48,47,46,44]
min_t = [20,25,24,31,27,26,29]
average_t = [35,36,37,38,34,36,31]
plt.xlabel("Days")
plt.ylabel("Temprature")
plt.title("Weather")
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,average_t,label="Average")
#use "best" for automatically decission
plt.legend(loc="best",shadow=True,fontsize="large")
plt.grid()
plt.show()
```

```python
import numpy as np
Company = ["AMZN","FB","GOOGLE"]
revenue = [23,21,22]
profit= [6,4,5]
xpos = np.arange(len(Company))
xpos
```

```python
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,Company)
plt.xlabel("Company")
plt.ylabel("Revenue $")
plt.title("Revenue sheet")
plt.bar(xpos-0.2,revenue,color="pink",label="Revenue",width=0.3)
plt.bar(xpos+0.2,profit,label="Profit",width=0.3)
plt.grid()
plt.legend(loc="best")
```

```python
plt.yticks(xpos,Company)
plt.ylabel("Company")
plt.xlabel("Revenue $")
plt.title("Revenue sheet")
plt.barh(xpos-0.2,revenue,color="pink",label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")
plt.grid()
plt.legend(loc="best")
```

```python
man_sugar_level = [80,85,90,95,97,100,105,120,123,124,137,150]
women_sugar_level = [60,65,80,85,87,90,95,92,93,94,97,99]
plt.hist([man_sugar_level,women_sugar_level],bins=[60,85,95,100,125,150],rwidth=0.94,color=["g","y"],label=["Men","Women"],orientation="horizontal")
plt.legend()
```

```python
expense = [10000,4000,5000,3000]
expense_label= ["Food","Rent","Dressing","Fare"]
plt.axis("equal")
plt.pie(expense,labels=expense_label,radius=1,autopct="%0.2f%%",shadow=True,explode=[0,0.3,0,0],startangle=95)
plt.savefig("piechart.png",bbox_inches="tight",pad_inches=2,transparent=True)
plt.savefig("piechart2.pdf",bbox_inches="tight",pad_inches=2,transparent=True)
```
